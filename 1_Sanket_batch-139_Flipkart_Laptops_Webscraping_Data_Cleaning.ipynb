```python
import requests as rq
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as bs
import re
```

###### Extracting the Details From Website of Flipkart
##### Product Name, Ratings, Price, Specifications, Page Number and Discount

```python
productname = []
stars = []
price = []
Features = []
no_of_rat = []
no_of_rev = []
pageNo=[]
discounted=[]

for n in range(1,20):
    URL = 'https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={}'.format(n)
    page=rq.get(URL)
    pagecontent = page.text
    soup = bs(pagecontent)
    # Accessing the Div element where all the product details are Stored
        # Product name
        #Price
        #ratings
        #etc
    for x in soup.find_all('div',attrs = {'class':'_3pLy-c row'}):
        
        #This extracts the product name from the webiste
        pname = x.find('div',attrs = {'class':'_4rR01T'})
        #This will extract the Cost price
        cost = x.find('div',attrs ={'class':'_1_WHN1'})
        #This will extract the Stars given to the laptop
        star = x.find('div',attrs = {'class':'_3LWZlK'})
        #This will extract the discount
        discount = x.find('div',attrs = {'class':'_3Ay6Sb'})
        
        #This will extract the Features
        features = ""
        fea = x.find("ul",attrs = {"class":"_1xgFaf"})
        for each_feature in fea:
            features = features + "||" + each_feature.text 
        

        #This will extract the Number of Reviews and Ratings 
        no_rat_rev = x.find('span',attrs = {'class':'_2_R_DZ'})
        
        #This will create a list of all details for dataFrame. If the scraped text happens to be none then Nan will be appended
        if pname is None:
            productname.append(np.NaN)
        else:
            productname.append(pname.text)
        
        if cost is None:
            price.append(np.NaN)
        else:
            price.append(cost.text)
            
        if star is None:
            stars.append(np.NaN)
        else:
            stars.append(star.text)
            
        if fea is None:
            Features.append(np.NaN)
        else:
            Features.append(features)
            
        if discount is None:
            discounted.append(np.NaN)
        else:
            discounted.append(discount.text.strip("% off"))
            
        if(no_rat_rev != None):
            no_rat = no_rat_rev.text.split('\xa0&\xa0')[0]
            no_rev = no_rat_rev.text.split('\xa0&\xa0')[1]
            no_of_rat.append(no_rat)
            no_of_rev.append(no_rev)
        else:
            no_of_rat.append(np.NaN)
            no_of_rev.append(np.NaN)

            
        pageNo.append(n)
```

```python
# laptop_ratings = soup.find('span', attrs = {'class': '_2_R_DZ'})
# laptop_ratings.text
```

###### Converting the Scraped Data Into the DataFrame and Then Saving it into csv file 

```python
flipKart=pd.DataFrame({"Product":productname, "Price":price, "Stars":stars, "Ratings":no_of_rat, "Reviews":no_of_rev, "Features":Features, "PageNumber":pageNo, "Discount":discounted})
flipKart.to_csv("Webscraping_flipkart_laptop_data.csv",index=False)
flipKart
```

#### Lets Add Some columns to our dataset so that we can make precise analysis

#### Extracting Brand,ProcessorName,RAM size,Storage,OS,Processor , Generation,GC size and Brand via regex

```python
df["Brand"]=df["Product"].apply(lambda x:re.compile("^\w+").findall(x)[0])
```

```python
re_list_features = r'[|]{2}\b([^|]+)'
df['list_features'] = df['Features'].apply(lambda x: re.findall(re_list_features , x))
```

```python
df['Specifications'] = np.NaN
```

```python
for i in range(len(df)):
    for feature in re.findall(re_list_features , df.loc[i,'Features']):
        if 'Processor' in feature:
            df.loc[i,'Specifications'] = feature
```

```python
df["Ratings"] = df["Ratings"].apply(lambda x : str(x).replace(',','').split(' R')[0]).astype('float')
df["Reviews"] = df["Reviews"].apply(lambda x : str(x).replace(',','').split(' R')[0]).astype('float')
```

```python
df.head(2)
```

```python
df.loc[46,'Specifications']
```

```python
df["ProcessorBrand"]=df["Specifications"].apply(lambda x:" ".join(re.compile("(?:Intel|AMD|Apple)").findall(x)[:2]))
df["RAM"]=df["Product"].apply(lambda x:" ".join(re.compile("\d{1,2} GB\/|\d{1,2}GB\/|\d{1,2} GB \+ \d{1,2} GB").findall(x))).str.strip("/")
df["SSD/HDD Size"]=df["Product"].apply(lambda x:" ".join(re.compile("\d TB [SSD|HDD]+|\d{3} GB [SSD|HDD]+").findall(x)))
df["OS"]=df["Product"].apply(lambda x:" ".join(re.compile("Windows \d{1,2}|Mac OS").findall(x)))
df["GraphicCardSize"]=df["Product"].apply(lambda x:" ".join(re.compile("\d{1,2} GB Graphics*").findall(x))).str.strip("Graphics")
df["GraphicCardBrand"]=df["Product"].apply(lambda x:" ".join(re.compile("Graphics/(.*)\)?").findall(x)))
```

```python
df["ProcessorName"]=df["Product"].apply(lambda x:"".join(re.compile("(?:Ryzen) ?\d?|(?:Pentium|AthlonCore|Celeron|Core)|(?: i\d)").findall(x)))
```

```python
df["ProcessorGeneration"]=df["Specifications"].apply(lambda x:"".join(re.compile("(?:Ryzen \d+)|(?:\d{1,2}th Gen+)").findall(x)))
```

###### For Graphic Card Brand we see there are some nulls present and most of them are For NVIDIA, so lets substitute them properly with name as NVIDIA

```python
df.loc[(df["GraphicCardBrand"].str.startswith("N")),"GraphicCardBrand"]="NVIDIA"
```

```python
df["ProcessorName"].replace("PentiumCore","Pentium",inplace=True)
df.loc[(df["Product"]=="Lenovo B50-70 Notebook (4th Gen Ci5/ 8GB/ 1TB/ Win8/ 2GB Graph) (59-427747)"),"ProcessorName"]="Core i5"
df["ProcessorName"].replace("Ryzen 5CoreRyzen 5","Ryzen 5Core",inplace=True)
df["ProcessorName"].replace("CeleronCore","Celero",inplace=True)
df["ProcessorName"].replace("Ryzen 7CoreRyzen","Ryzen 7Core",inplace=True)
df["ProcessorName"].replace("","np.NaN",inplace=True)
df["ProcessorBrand"].replace("",np.NaN,inplace=True)
```

```python
df["ProcessorGeneration"].replace("Ryzen 95th Gen","Ryzen 9",inplace=True)
df["ProcessorGeneration"].replace("Ryzen 55th Gen","Ryzen 5",inplace=True)
df["ProcessorGeneration"].replace("Ryzen 74th Gen","Ryzen 7",inplace=True)
df["ProcessorGeneration"].replace("",np.NaN,inplace=True)
```
